Wood-Saxon

```python
def WoodSaxon(atom):
    pr = lambda r: (p0*(1+(w*r**2)/(R**2)))/(1+e**((r-R)/a))
    if atom == "C" or atom == 12:
        R = 2.47
        a = 0.
        w = 0.
    elif atom == "O" or atom == 16:
        R = 2.608
        a = 0.513
        w = -0.051
    elif atom == "Al" or atom == 27:
        R = 3.07
        a = 0.519
        w = 0.
    elif atom == "S" or atom == 32:
        R = 3.458
        a = 0.61
        w = 0.
    elif atom == "Ca" or atom == 40:
        R = 3.76
        a = 0.586
        w = -0.161
    elif atom == "Ni" or atom == 58:
        R = 4.309
        a = 0.516
        w = -0.1308
    elif atom == "Cu" or atom == 63:
        R = 4.2
        a = 0.596
        w = 0.
    elif atom == "W" or atom == 186:
        R = 6.51
        a = 0.535
        w = 0.
    elif atom == "Au" or atom == 197:
        R = 6.38
        a = 0.535
        w = 0.
    elif atom == "Pb" or atom == 208:
        R = 6.68
        a = 0.546
        w = 0.
    elif atom == "U" or atom == 238:
        R = 6.68
        a = 0.6
        w = 0.
    p0 = 1.
    r = np.arange(0.,2*R,0.001)
    return pr(r)
```

```python
r_O = np.arange(0.,2*2.608,0.001)
plot(r_O,WoodSaxon("O"),label="Density Plot of Oxygen")

r_Cu = np.arange(0.,2*4.2,0.001)
plot(r_Cu,WoodSaxon("Cu"),label="Density Plot of Copper")

r_Pb = np.arange(0.,2*6.68,0.001)
plot(r_Pb,WoodSaxon("Pb"),label="Density Plot of Lead")

legend(loc=0)
title("Density Plots of Select Elements",fontsize=24)
xlabel("Radius (fm)",fontsize=18)
ylabel("Wood-Saxon Density",fontsize=18)

show()
```
