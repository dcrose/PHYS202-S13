###Example 1: Analytic Inverse CDF Transform

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label="True Distribution",lw=2)
plot(x,F(x),label="Cumulative Function",lw=2)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=2)
show()
```

```python
#Number of samples
N = 100000

#N random numbers between 0 and 1
z = random.random_sample(N)

#Inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)

#Sampled x values
xsamples = Finv(z)

#How do they compare to the original?
hist(xsamples,50,normed=True,label="Sampled Distribution")
plot(x,f(x),label="True Distribution",color="red",lw=2)

xlim(0,pi)
ylim(0,0.7)

legend(loc=1)
show()
```

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, 
    on the interval defined by the array x, and 
    returns an array of N sampled values that are 
    statistically the same as the input data."""
    z = random.random_sample(N);
    c = cumsum(prob)
    
    a = sum(prob)
    c = c/a
    s = c.searchsorted(z)
    thing = x[s]
    return thing
```

```python
x = arange(0.,5.01,0.01)
t = h(x)
s = distribute1D(x,t,10000)
hist(s,50,normed=True,label="Sampled Distribution")
show()
```
