Cumulative Distribution Function
-
###Example 1: Analytic Inverse CDF Transform

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.
    
plot(x,f(x),label="True Distribution",lw=2)
plot(x,F(x),label="Cumulative Function",lw=2)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=2)
show()
```

```python
#Number of samples
N = 100000

#N random numbers between 0 and 1
z = random.random_sample(N)

#Inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)

#Sampled x values
xsamples = Finv(z)

#How do they compare to the original?
hist(xsamples,50,normed=True,label="Sampled Distribution")
plot(x,f(x),label="True Distribution",color="red",lw=2)

xlim(0,pi)
ylim(0,0.7)

legend(loc=1)
show()
```

###Exercise 1: Exponential function - analytic

##$f(x) = \lambda e^{-\lambda x}$

##$F(x) = \int_{-∞}^x f(x) dx = -e^{-\lambda x}$

```python
N = 100000
z = random.random_sample(N)
l = 1.
g = lambda x: l*e**(-l*x)
G = lambda x: -e**(-l*x)+1

plot(x,g(x),label="True Distribution",lw=2)
plot(x,G(x),label="Cumulative Function",lw=2)

legend(loc=0)
```

```python
Ginv = lambda z: -(log(-z+1))/l
x1samples = Ginv(z)
hist(x1samples,150,normed=True,label="Sampled Distribution")
plot(x,g(x),label="True Distribution",color="red",lw=2)

xlim(0,5)
ylim(0,1.1)
legend(loc=0)
show()
```
