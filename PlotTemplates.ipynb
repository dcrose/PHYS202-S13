```python
y = arange(-5,6,0.1) # Creates an array with  a range from -5 to 5.
E = cosh(y)      # Names the variable E to be the hyperbolic cosine of y.
p = sinh(y)      # Defines a  variable p as the hyperbolic sine of y.
z = abs(p)           # Calls the absulolute value of p.
```

```python
plot(y,z,color="red",label="|p|") # Creates a plot of the absolute value of p vs. y.
plot(y,E,color="blue",label="E")  # Creates a plit of E vs. y.
xlabel("y",fontsize=15)           # Creates an x-label.
ylabel("E , |p|",fontsize=15)     # Creates a y-label.
title("Plot of E , |p| vs. y",fontsize = 20) # Creates a title.
legend(loc = "upper left")        # Initializes the legend.
ax = gca()
ax.spines["right"].set_color("none") # Removes the right border.
ax.spines["top"].set_color("none")   # Removes the top border.
ax.spines["left"].set_color("none")  # Removes the left border.
ax.yaxis.set_ticks_position("none")
ax.xaxis.set_ticks_position("bottom")
grid(True)
```
