```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, 
    on the interval defined by the array x, and 
    returns an array of N sampled values that are 
    statistically the same as the input data."""
    z = random.random_sample(N);
    c = cumsum(prob)
    
    a = sum(prob)
    c = c/a
    s = c.searchsorted(z)
    thing = x[s]
    return thing

d = 15*10**(-6) #meters
a = 10*10**(-6) #meters
L = 1. #meters
A = 1.
wavelength = 632.8*10**(-9) #meters
f = lambda x: A**4*((sinc(a*x/(wavelength*L)))**2 * cos(2*pi*d*x/(wavelength*L)))**2
dx = 0.001
x = np.arange(-0.2,0.21,dx)
plot(x,f(x))
show()
```
