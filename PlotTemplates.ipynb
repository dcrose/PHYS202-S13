Line plots
-

```python
y = arange(-5,6,0.1) # Creates an array with  a range from -5 to 5.
E = cosh(y)      # Names the variable E to be the hyperbolic cosine of y.
p = sinh(y)      # Defines a  variable p as the hyperbolic sine of y.
#z = abs(p)           # Calls the absulolute value of p.
```

```python
plot(y,p,color="red",label="|p|") # Creates a plot of the absolute value of p vs. y.

plot(y,E,color="blue",label="E",linestyle="--",lw=2)  # Creates a plit of E vs. y.

xlabel("y",fontsize=15,position=(.83, 0))           # Creates an x-label.

ylabel("E , |p|",fontsize=15,position=(0, .8))     # Creates a y-label.

title("Plot of E , |p| vs. y",fontsize = 20) # Creates a title.

legend(loc = "lower right")        # Initializes the legend.

axis([-2*pi,2*pi,-50,50])          # Sets the axis limits.

ax = gca()

ax.spines["right"].set_position("center") # moves the right border.

ax.spines["bottom"].set_position("center")  # moves the bottom border.

ax.spines["left"].set_position("center")  # moves the left border.

ax.spines["top"].set_position("center")   # moves the top border

ax.yaxis.set_ticks_position("left")

ax.xaxis.set_ticks_position("bottom")

grid(True)
```

Subplots:
-
How to show multiple, separate graphs at the same time.

```python
figure(2, figsize=(10.5,5)) #make new figure with index 2

subplot(1,2,1) #In a 1-row, 2-column grid, sets the current subplot to position 1
plot(y, E, 'r-', lw=2)
#insert nice formatting here, as above
#any formatting done here will only apply to this subplot:
title("Subplot 1")

subplot(122) #In a 1-row, 2-column grid, sets the current subplot to position 2 (shorthand form)
plot(y, p, 'r-', lw=2)
#insert nice formatting here, as above
#any formatting done here will only apply to this subplot:
title("Subplot 2")

subplot(121) #We can switch the active subplot at any time if we need to
axis([-5,5,-1,10])

show()
```
