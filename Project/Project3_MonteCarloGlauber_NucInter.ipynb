```python
# Now to interpolate this thing.

def interacting(E):
    from scipy.interpolate import interp1d
    cs_total = interp1d(PLABz,SIGw)(E)
    cs_elas = interp1d(xex,cf)(E)
    cs_inelas = cs_total - cs_elas
    distance = ((cs_inelas*10**(-1))/pi)**(1./2.)
    return distance
interacting(10)
```

Next, we apply the Wood-Saxon probability distribution function to the chosen atoms.

At this point, all distances are in fm.

```python
import random

def WoodSaxon(atom):
    prob = lambda r: (p0*(1+(w*r**2)/(R**2)))/(1+e**((r-R)/a)) # Wood-Saxon probability distribution.
    
    # These are our chosen atoms.
    if atom == "C" or atom == 12:
        n = 12.
        R = 2.47
        a = 0.
        w = 0.
    elif atom == "O" or atom == 16:
        n = 16.
        R = 2.608
        a = 0.513
        w = -0.051
    elif atom == "Al" or atom == 27:
        n = 27.
        R = 3.07
        a = 0.519
        w = 0.
    elif atom == "S" or atom == 32:
        n = 32.
        R = 3.458
        a = 0.61
        w = 0.
    elif atom == "Ca" or atom == 40:
        n = 40.
        R = 3.76
        a = 0.586
        w = -0.161
    elif atom == "Ni" or atom == 58:
        n = 58.
        R = 4.309
        a = 0.516
        w = -0.1308
    elif atom == "Cu" or atom == 63:
        n = 63.
        R = 4.2
        a = 0.596
        w = 0.
    elif atom == "W" or atom == 186:
        n = 186.
        R = 6.51
        a = 0.535
        w = 0.
    elif atom == "Au" or atom == 197:
        n = 197.
        R = 6.38
        a = 0.535
        w = 0.
    elif atom == "Pb" or atom == 208:
        n = 208.
        R = 6.68
        a = 0.546
        w = 0.
    elif atom == "U" or atom == 238:
        n = 238.
        R = 6.68
        a = 0.6
        w = 0.
    p0 = 1.
    r = np.arange(0.,2*R,0.001) # Returns the radius values!
    d = random.uniform(0.,1.5)*R # Returns a random interacting distance!
    return (prob(r),n,d,r,R)
```

Now we want to create a function to generate nucleons and gives them values for theta, the interacting angle, and the radius r.

```python
def r_rad(atom):
    t = numpy.random.random_sample(WoodSaxon(atom)[1]) # The number of samples required depends on the atom type.
    s = cumsum(WoodSaxon(atom)[0]) # Continuously sum it all up.
    a = sum(WoodSaxon(atom)[0]) # Sum the array.
    s = s/a # Normalize.
    c = s.searchsorted(t) # Sort.
    r = WoodSaxon(atom)[3] # Gives the radius of the atom we want.
    radDist = r[c] # Gives a radius value according to the random sample and probability distribution.
    theta = zeros(len(radDist)) # The angle.
    count = 0 # Start the counter.
    for i in range(0,len(radDist)): # For creating the random angles.
        theta[count] = random.uniform(0,2*pi)
        count += 1
    return (radDist,theta,WoodSaxon(atom)[0],WoodSaxon(atom)[1],WoodSaxon(atom)[2],WoodSaxon(atom)[3],WoodSaxon(atom)[4])
```

```python
# Create a function that takes in two atoms, and gives the plots we want.
def collision(a1,a2,E):
    """Takes in two atoms, a1 and a2, and their energy E, and returns the collision plots!"""
    # This will call the function r_rad on the atoms we want to generate the values we need.
    a1_dist = r_rad(a1)
    a2_dist = r_rad(a2)
    # For this, a_dist[4] would be the d values,
    # a_dist[0] the r values,
    # and a_dist[1] the theta values.
    
    # Now to plot each:
    plot(a1_dist[0]*cos(a1_dist[1])-a1_dist[4],a1_dist[0]*sin(a1_dist[1]),linestyle="*",color="red",label=a1 +" a/a1")
    
    plot((a2_dist[0]*cos(a2_dist[1]))+a2_dist[4],a2_dist[0]*sin(a2_dist[1]),linestyle="*",color="blue",label=a2+" a/a2")
    
    # We will want to change the r and theta values of these to x,y coordinates:
    a1_dist_co = ((a1_dist[0]*cos(a1_dist[1])-a1_dist[4]),(a1_dist[0]*sin(a1_dist[1])))
    a2_dist_co = ((a2_dist[0]*cos(a2_dist[1])+a2_dist[4]),(a2_dist[0]*sin(a2_dist[1])))
    
    dist = interacting(E) # Recall that this is used to find the radius for a given energy.
    
    # Assign the coordinates to individual variables:
    a1_x = a1_dist_co[0]
    a1_y = a1_dist_co[1]
    a2_x = a2_dist_co[0]
    a2_y = a2_dist_co[1]
    
    # Now we want new arrays to add the collision coordinates to:
    collision_a1_x = []
    collision_a1_y = []
    collision_a2_x = []
    collision_a2_y = []
    
    #Next, for loops to add the coordinates to the empty arrays:
    for s in range(0,len(a1_x)):
        for t in range(0,len(a2_x)):
            # This will find the distance between the two atoms
            distance = ((a1_x[s]-a2_x[t])**2+(a1_y[s]-a2_y[t])**2)**(1./2.)
            if distance < 2*dist:
                count += 1
                # Append the empty arrays with the coordinate data
                collision_a1_x.append(a1_x[s])
                collision_a1_y.append(a1_y[s])
                collision_a2_x.append(a2_x[t])
                collision_a2_y.append(a2_y[t])
    #part_count = 0 # Starts the count for the number of participants at zero.
    #could not get the participant count to work.

    plot(collision_a1_x,collision_a1_y,linestyle="*",color="green",label="Nucleons from atom 1")
    plot(collision_a2_x,collision_a2_y,linestyle="*",color="magenta",label="Nucleons from atom 2")
    
    ax = gca()
    #ax.annotate() # No count to annotate in...
    ax.annotate("The impact parameter d = "+str(a1_dist[4]+a2_dist[4]),xy=((a2_dist[4]-a1_dist[4])/2-3,-11),fontsize=14)
    ax.annotate(str(len(a1_x)+len(a2_x))+" nucleons are participant",xy=(-7,19),fontsize=14)
    
    # Now cicles!
    import matplotlib.pyplot as pyplot
    Circle_1 = pyplot.Circle((-a1_dist[4],0),a1_dist[6],color="blue",clip_on=False,fill=False)
    Circle_2 = pyplot.Circle((a2_dist[4],0),a2_dist[6],color="red",clip_on=False,fill=False)
    fig=pyplot.gcf()
    fig.gca().add_artist(Circle_1)
    fig.gca().arr_artist(Circle_2)
    xlabel("x Position (fm)",fontsize=20)
    ylabel("y Position (fm)",fontsize=20)
    xlim(-20,20)
    ylim(-20,20)
    legend(loc=0)
```
