```python
f = lambda t: ((1+t)/(t-1))
g = lambda t: e**(-t)

t = arange(-5,5.01,0.001)

plot(t,f(t),color="blue",label="(1+t)/(t-1)")
plot(t,g(t),color="red",linestyle=":",label="exp(-t)")

axis([-5,5,-10,10])

ax = gca()
ax.annotate("discontinuous here",xy=(1.2,-3),fontsize=20)

xlabel("time (seconds)",fontsize=18)
ylabel("f(t), g(t)",fontsize=18)
legend(loc="lower left")
show()
```
