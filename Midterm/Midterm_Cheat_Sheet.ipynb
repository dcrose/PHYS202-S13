```python
from Differentiators import *
```

```python
x1 = arange(0,10,0.01)
y1 = x1**5
dydx_exact = 5*x1**4
dydx1 = finiteDifference(x1,y1)
dydx2 = fourPtFiniteDiff(x1,y1)
```

```python
figure(figsize=(10,10))

semilogy(x1, 100*(abs(dydx2-dydx_exact))/dydx_exact, label = "4 Point Centered Differential")
semilogy(x1, 100*(abs(dydx1-dydx_exact))/dydx_exact, label = "2 Point Centered Differential")
xlabel("$x$", fontsize=18)
ylabel('% Error', fontsize=18)
legend(loc="upper right")
title('Accuracy of Derivatives of $x^5$', fontsize=18)

show()
```
