##Example 1:

```python
f = lambda x: x**(0.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r"$x^{1/2} + \sin(x)$",fontsize=20)
xlabel(r"$x$",fontsize=20)
xlim(0.,15.)

show()
```

```python
xmin = 1.7
xmax = 12.4

# Find the maximum value of f within the boundary to set ymax
subx = x[logical_and((x > xmin),(x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

# Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

# Plot the box
plot([xmin,xmin],[ymin,ymax],color="k",linestyle="--") # Vertical left
plot([xmax,xmax],[ymin,ymax],color="k",linestyle="--") # Vertical right
plot([xmin,xmax],[ymax,ymax],color="k",linestyle="--") # Horizontal top

show()
```

```python
#random_sample?

N = 10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

# Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],".")

# Plot the box
plot([xmin,xmin],[ymin,ymax],color="k",linestyle="--") # Vertical left
plot([xmax,xmax],[ymin,ymax],color="k",linestyle="--") # Vertical right
plot([xmin,xmax],[ymax,ymax],color="k",linestyle="--") # Horizontal top
```

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

```python
# Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],".")

# Show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],"r.")

# Plot the box
plot([xmin,xmin],[ymin,ymax],color="k",linestyle="--") # Vertical left
plot([xmax,xmax],[ymin,ymax],color="k",linestyle="--") # Vertical right
plot([xmin,xmax],[ymax,ymax],color="k",linestyle="--") # Horizontal top

show()
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

```python
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```

Write code to compute the integral of the function

from example 1 using the sample mean method for N = 10000

```python
f = lambda x: x**(0.5) + sin(x)
N = 10000.
a = xmin
b = xmax
x_random = xmin + (xmax-xmin)*random_sample(N)
total = sum(f(x_random))
I2 = (b-a)*(total/N)
print I2
```

##Example 3: Electric potential of a plane with constant charge density

Let's consider another example, the potential from a plane of charge as shown in the figure (notes).

```python
#add the Week3 directory to search path
import sys
sys.path.insert(0,"../Week4")

# Import out point potential function
from Electrostatics import *

# Define our variables
sigma = 1e-6;                 # Charge density is 1.0 microC/m^2
N = 1e6;
x0 = 2.;                      # P(x0,y0)
y0 = 3.;
xi = -1. + 2.001*random_sample(N);  # -1 < x < 1 integration limits
yi = 0. + 2.001*random_sample(N);   # 0 < y < 2 integration limits

#electric potential at P from a chunk of charge at xi, yi
Vi = pointPotential(x0,y0,sigma,xi,yi);

# Integrate Vi from chuncks of charge in x and y directions
Vtot = (2.*2./N)*sum(Vi)    # range is 2 in x and y

print Vtot
```
