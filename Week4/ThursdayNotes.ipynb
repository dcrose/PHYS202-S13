Using Boolean masks to find data
=

```python
theta = arange(0.,2*pi,pi/100)
```

```python
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
#print mask
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
```

```python
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```

```python
thetaNew = arange(0.,2*pi+0.1,pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta),thetaNew,sinNew, "r.",markeredgewidth = 1)
show()
```

```python
plot(thetaNew,sinNew - sin(thetaNew))
ylabel("Difference between interp and exact value")
show()
```

```python
sinNewC = interp1d(theta,sin(theta),kind = "cubic")(thetaNew)#Cubic sin interpolation
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew),thetaNew,sinNewC-sin(thetaNew), "r")
subplot(2,1,2)
plot(thetaNew,sinNewC-sin(thetaNew),"r")
show()
```

Differentiation
=

```python
x = arange(0.,10.01,0.1)
y = x**5
```

```python
dy = diff(y) # y(n+1) - y(n)
#help(diff)
print len(y), len(dy)
dx = diff(x)
dydx = (dy/dx)
print len(dydx)
```

```python
#plot(x,dydx) #doesnt work because the ranges of x and dydx are not the same. They are off by one.
```

```python
# Specify the size of dy ahead because diff return an array of n-1 elements
dydx1 = zeros(y.shape,float) #we know it will be this size
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
```

```python
# Calculate dydx by center differencing using array slices
dydx2 = zeros(y.shape,float)
dydx2[1:-1] = (y[2:] - y[:-2])/(x[2:] - x[:-2]) #center difference
dydx2[0] = (y[1] - y[0])/(x[1] - x[0]) #forward difference
dydx2[-1] = (y[-1] - y[-2])/(x[-1] - x[-2]) #backward difference
```

```python
dydxExact = 5*x**4
```

```python
plot(x,dydx1,label="forward difference")
plot(x,dydx2,label="center difference")
plot(x,dydxExact,label="exact")
legend(loc="upper left")
show()
#help(plot)
```

```python
percentError1 = 100*abs(dydx1[1:] - dydxExact[1:])/dydxExact[1:]
percentError2 = 100*abs(dydx2[1:] - dydxExact[1:])/dydxExact[1:]
# Use a semilog y-axis to see the variation better
semilogy(x[1:],percentError1,label="forward difference")
semilogy(x[1:],percentError2,label="center difference")
ylabel("percent error with dydxExact")
legend(loc="upper right")
show()
```
