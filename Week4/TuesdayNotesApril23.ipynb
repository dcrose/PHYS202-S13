Electric Dipole Potential
=

#$V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
# Create an x-y grid covering the plane -2<x<2 and -2<y<2
dx = 0.06
dy = 0.06
x,y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
print x.ndim
print y.shape
```

```python
k = 1; q = 1; d = 1
V=(k*q/(x**2+(y-d)**2)**(1/2.))-(k*q/(x**2+(y+d)**2)**(1/2.))
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = fig.gca(projection="3d")
p = ax.plot_surface(x,y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.prism_r)



cb = fig.colorbar(p,shrink=0.5)
cset = ax.contour(x,y,V,zdir="z",offset=-40.,cmap=cm.prism_r)
cset = ax.contour(x,y,V,zdir="x",offset=-2.,cmap=cm.prism_r)
cset = ax.contour(x,y,V,zdir="y",offset=-2.,cmap=cm.prism_r)

```

```python
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])
contour(x,y,V,levels,cmap=cm.jet)
show()
```

```python

def pointPotential(x,y,q,posx,posy):
    k = 8.98755*10**9; q = 1.6e-19
    Vpoint = (k*q/((x-posx)**2+(y-posy)**2)**(1/2.))
    return Vpoint
print pointPotential(1.,1.,1.6e-19,0.,0.)

```

###Part b)

```python
def dipolePotential(x,y,q,d):
    k = 8.98755*10**9; q = 1.6e-19
    Vdipole = (k*q/(x**2+(y-(d/2))**2)**(1/2.))-(k*q/(x**2+(y+(d/2))**2)**(1/2.))
    return Vdipole
print dipolePotential(1.,1.,q,3.)
```

###Part c)
