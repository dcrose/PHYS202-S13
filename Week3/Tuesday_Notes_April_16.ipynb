Create a very simple plot.

```python
import numpy as np
t = np.arange(0.0, 1.0+0.01, 0.01)
c = np.cos(4*pi*t)
s = np.sin(4*pi*t)
```

```python
print pi
```

```python
plot(t,c, color = "blue", linewidth=2.5,label="$cos(4{\pi}t)$") #Plots the function in terms of t on the x and c on the y, changes the color to blue, the line size to 2.5, and labels it cos(etc).
plot(t,s, color = "red",linewidth=2.5,linestyle=":",label="$sin(4{\pi}t)$") # Does something similar for the s function. Also makes it a dotted line with the linestyle command.
xlabel("time (s)", fontsize = 15) # Creates a label for the x axis.
ylabel("voltage (mV)", fontsize = 15) # Creates a label for the y axis.
title("A simple plot", fontsize = 20) # Creates a  title for the graph of size 20.
grid(True)
legend(loc="lower left")
savefig("Simple_plot.pdf")
```
